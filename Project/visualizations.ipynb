```python
import matplotlib.pyplot as plt
import matplotlib.colors as clrs
import numpy as np
```

```python
#plot time vs. recall for different number of clusters
store_time_pqs= [0.09034848, 0.02689147, 0.02662134, 0.02627778, 0.02639198, 0.02782416,
0.02777719, 0.03004432, 0.03574872, 0.05114985]
store_recall_pqs= [0.843, 0.59,  0.477, 0.306, 0.264, 0.213, 0.206, 0.209, 0.238, 0.298]
nlist_values= [2**i for i in range(10)]
```

```python
def plot_recall_time(recall,time,nlists, title):
    plt.figure(figsize=(10,7))
    plt.scatter(recall, time, c=nlists, norm=clrs.LogNorm())
    plt.xlabel('Recall@5')
    plt.ylabel('Time')
    plt.title(title)
    plt.xlim(0,1)
    plt.ylim(0,.15)
    cbar = plt.colorbar()
    cbar.set_label('Number of Clusters')
```

```python
plot_recall_time(store_recall_pqs, store_time_pqs, nlist_values,'Recall vs. Time for normal distribution, n=1000,d=40')
```
