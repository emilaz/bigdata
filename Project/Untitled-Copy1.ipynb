```python
from lsh import *
import pandas as pd
```

```python
datapoints = 10000
```

```python
def get_normal():
    df = pd.DataFrame(columns=['Variance','k','L','LSH Time','Error'])
    for var in np.linspace(1,101,10): #different variances
        print('Variance', var)
        A = np.random.normal(255 / 2, var, (datapoints, 400)) #draw from dist
        queries = np.random.choice(datapoints,20)
        k_range = range(0,41,5)
        for k in k_range:
            for L in range(0,41):
                if k*L != 40:
                    continue
                print('k={}, L={}'.format(k,L))
                lin_time, lsh_time, err = get_lin_and_lsh_times(A, queries, L, k, 3)
                df.loc[len(df.index)] = [var,k,L,lsh_time,err]
    return df

def get_uniform():
    df = pd.DataFrame(columns=['k','L','LSH Time','Error'])
    A = np.random.uniform(0,255, (datapoints, 400)) #draw from dist
    queries = np.random.choice(datapoints,20)
    k_range = range(0,41,5)
    for k in k_range:
        for L in range(0,41):
            if k*L != 40:
                continue
            print('k={}, L={}'.format(k,L))
            lin_time, lsh_time, err = get_lin_and_lsh_times(A, queries, L, k, 3)
            print('Error',err)
            df.loc[len(df.index)] = [k,L,lsh_time,err]
    return df

```

```python
df = get_normal()
```

```python
df_uni = get_uniform()
```

```python
df.to_hdf('res_grid_search_normal.hdf', key='df', mode='w')
```

```python
df_uni.to_hdf('res_grid_search_uniform.hdf',key='df',mode='w')
```

```python
by_kl = df.groupby(['k','L'])[['Variance','LSH Time','Error']].apply(lambda x: x.values.tolist())
```

```python
best_uni_error = df_uni.iloc[df_uni['Error'].values.argmin()]
best_uni_time = df_uni.iloc[df_uni['LSH Time'].values.argmin()]
```

```python
for k,L in by_kl.index:
    tuples = by_kl[(k,L)]
    var = [t[0] for t in tuples]
    time =[t[1] for t in tuples]
    err = [t[2] for t in tuples]
    plt.plot(var,time,label ='k= {}, L={}'.format(int(k),int(L)))
    plt.title('Different choices of k and L over increasing variance')
    plt.xlabel('Variance')
    plt.ylabel('Time')
plt.legend()
plt.show()
    
for k,L in by_kl.index:
    tuples = by_kl[(k,L)]
    var = [t[0] for t in tuples]
    time =[t[1] for t in tuples]
    err = [t[2] for t in tuples]
    plt.plot(var,err,label ='k= {}, L={}'.format(int(k),int(L)))
    plt.title('Different choices of k and L over increasing variance')
    plt.xlabel('Variance')
    plt.ylabel('Error')
plt.legend()
plt.show()
```

```python
gauss_time_best_uni_time = [b[1] for b in by_kl[(best_uni_time['k'],best_uni_time['L'])]]
gauss_time_best_uni_err = [b[1] for b in by_kl[(best_uni_error['k'],best_uni_error['L'])]]
gauss_err_best_uni_time = [b[2] for b in by_kl[(best_uni_time['k'],best_uni_time['L'])]]
gauss_err_best_uni_err = [b[2] for b in by_kl[(best_uni_error['k'],best_uni_error['L'])]]
var = [b[0] for b in by_kl[(best_uni_error['k'],best_uni_error['L'])]]
```

```python
plt.plot(var, gauss_time_best_uni_time, 'r', label = 'k= {}, L={}'.format(int(best_uni_time['k']),int(best_uni_time['L'])))
plt.plot(var, gauss_time_best_uni_err, 'g', label = 'k= {}, L={}'.format(int(best_uni_error['k']),int(best_uni_error['L'])))
plt.plot([best_uni_error['LSH Time']]*100,'g--',label = 'k= {}, L={}'.format(int(best_uni_error['k']),int(best_uni_error['L'])))
plt.plot([best_uni_time['LSH Time']]*100,'r--', label = 'k= {}, L={}'.format(int(best_uni_time['k']),int(best_uni_time['L'])))
plt.legend()
plt.xlabel('Variance')
plt.ylabel('Time')
plt.title('Uniform vs. Gaussian Distribution in terms of Time')

plt.show()

plt.plot(var, gauss_err_best_uni_time,'r', label = 'k= {}, L={}'.format(int(best_uni_time['k']),int(best_uni_time['L'])))
plt.plot(var, gauss_err_best_uni_err, 'g', label = 'k= {}, L={}'.format(int(best_uni_error['k']),int(best_uni_error['L'])))
plt.plot([best_uni_error['Error']]*100,'g--',label = 'k= {}, L={}'.format(int(best_uni_error['k']),int(best_uni_error['L'])))
plt.plot([best_uni_time['Error']]*100,'r--',label = 'k= {}, L={}'.format(int(best_uni_time['k']),int(best_uni_time['L'])))
plt.legend()
plt.xlabel('Variance')
plt.ylabel('Error')
plt.title('Uniform vs. Gaussian Distribution in terms of Error')
plt.show()
```
